```python
with open('data/The Great Gatsby/preproc.txt',"r", encoding="utf-8") as f:
    text = f.readlines()
    
print(text)
```

```python
# Preprocess
newlines = []
for i,line in enumerate(text):
    dont_include = False
    if line.endswith(" -\n"):
        newlines.append(line.removesuffix(" -\n") + text[i+1])
        dont_include = True
    else:
        if dont_include:
            dont_include = False
        else:
            newlines.append(line)
```

```python
with open('data/The Great Gatsby/preproc2.txt',"w", encoding="utf-8") as f:
    f.writelines(newlines)
```

```python
import pandas as pd
df = pd.read_csv("data/Hamlet/hamlets.csv")
text = df["text"][0]
print(text)
```

```python
from sentence_transformers import SentenceTransformer
import nltk

# SPLIT DOCS INTO SENTENCES
sentences = nltk.sent_tokenize("\n".join(text))

# GENERATE SENTENCE EMBEDDINGS
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
import pandas as pd

with open("generated_questions.txt","r",encoding="utf-8") as f:
    lines = f.readlines()
    
query_list = []
current_doc = ""
current_question = ""

for line in lines:
    if line.startswith("Document: "):
        current_doc = line.removeprefix("Document: ").removesuffix("\n")
    if line.startswith("Q"):
        current_question = line[4:].removesuffix("\n")
    if line.startswith("A"):
        query_list.append({
            "query":current_question,
            "answer": line[4:].removesuffix("\n"),
            "doc": current_doc              
        })
        
query_df = pd.DataFrame(query_list)
query_df.to_csv("gen_queries.csv")
```
